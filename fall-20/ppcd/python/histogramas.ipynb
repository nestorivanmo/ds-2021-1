```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
url = 'https://bit.ly/2cLzoxH'
brecha = pd.read_csv(url)
brecha.lifeExp.hist(bins=10)
```

```python
brecha.lifeExp.hist(bins=100)
```

```python
brecha.lifeExp.hist(bins=10)
plt.title('Histograma con pandas')
plt.xlabel('Esperanza de vida')
plt.ylabel('Frecuencia')
```

```python
brecha.gdpPercap.hist(bins=1000, grid=False)
plt.xlabel('Ingreso per cápita')
plt.ylabel('Frecuencia')
plt.xscale('log')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.displot(brecha.lifeExp, kde=True)
```

```python
sns.displot(brecha.lifeExp, bins=100, color='red')
plt.title('Histrograma con Seaborn')
plt.xlabel('Esperanza de vida')
plt.ylabel('Frecuencia')
```

```python
df = brecha[brecha.continent == 'Africa']
sns.distplot(df.lifeExp, kde=True, label='África')
df = brecha[brecha.continent == 'Americas']
sns.distplot(df.lifeExp, kde=True, label='América')

plt.title('Esperanza de vida en América y África')
plt.xlabel('Esperanza en años')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-white')
data = np.random.randn(1000)
plt.hist(data)
plt.show()
```

```python
x1 = data = np.random.normal(0,0.8,1000)
x2 = data = np.random.normal(-1.5,1,1000)
x3 = data = np.random.normal(2,1.6,1000)

formato = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)
plt.hist(x1, **formato)
plt.hist(x2, **formato)
plt.hist(x3, **formato)
plt.show()
```
