```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
url = 'https://bit.ly/2COHM14'
data = pd.read_csv(url)
data.head()
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(data)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
data_scaled.head()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title('Dendograma')
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit_predict(data_scaled)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data_scaled.Milk, data_scaled.Grocery, c=cluster.labels_)
```

```python
import math
P = [[8, 10], [3, 10.5], [7, 13.5], [5, 18], [5, 13], [6,9], [9, 11], [3, 18], [8.5, 12], [8, 16]]
c1 = [5,18]
c2 = [8, 10]

def find_norm(p, c, s=False):
    a = (p[0]-c[0])**2
    b = (p[1]-c[1])**2
    if s:
        return math.sqrt(a + b)
    return (a + b)

def find_dist(P, c):
    d = []
    for p in P:
        d.append(find_norm(p, c, s=False))
    return d
d1 = find_dist(P, c1)
d2 = find_dist(P, c2)
print(d1)
print(d2)

def find_groups(d1, d2):
    g1, g2 = [], []
    for i in range(len(d1)):
        if d1[i] >= d2[i]:
            g2.append(P[i])
        else:
            g1.append(P[i])
    return (g1, g2)

(g1, g2) = find_groups(d1, d2)
print(g1)
print(g2)

def find_centroids(g):
    s1 = 0
    s2 = 0
    for e in g:
        s1 += e[0]
        s2 += e[1]
    s1 /= len(g)
    s2 /= len(g)
    return [s1, s2]

c11 = find_centroids(g1)
c22 = find_centroids(g2)
print(c11)
print(c22)

```

```python
import numpy as np
import pandas as pd
from matplotlib import pylab as plt
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
data = pd.read_csv('http://dicyg.fi-c.unam.mx:8080/lalo/pypcd/presentaciones/xclara.csv')
print(data.shape)
data.head()
```

```python
f1 = data.V1.values
f2 = data.V2.values
plt.scatter(f1, f2, s=10)
plt.show()
```

```python
from sklearn.cluster import KMeans
X = np.array(list(zip(f1, f2)))
kmeans = KMeans(n_clusters=3)
kmeans = kmeans.fit(X)
y = kmeans.predict(X)
C_ = kmeans.cluster_centers_
C_
```

```python
fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:,1], c=y)
ax.scatter(C_[:,0], C_[:,1], marker='*', s=200, c='k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
plt.rcParams['figure.figsize'] = (12,9)
```

```python
X, y = make_blobs(n_samples=800, n_features=3, centers=4)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0], X[:,1], X[:,2])
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
labels = kmeans.predict(X)
C_ = kmeans.cluster_centers_
C_
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0], X[:, 1], X[:, 2], c=y)
ax.scatter(C_[:,0], C_[:,1], C_[:,2], marker='*', c='k', s=1000)
plt.show()
```
